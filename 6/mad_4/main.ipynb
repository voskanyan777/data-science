```python
y = df['Цена']
X = df.drop('Цена', axis=1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3,
                                                    random_state=0)
k_fold = KFold(n_splits=5, random_state=42, shuffle=True)
```

## Исследование зависимости качества прогназирования от настраиваемых параметров


## Begging

```python
n_estimators = [10, 50, 100]
max_samples = [0.5, 0.7, 1.0]
max_features = [0.5, 0.7, 1.0]

test_scores = [0] * 3
train_scores = [0] * 3
cv_scores = [0] * 3

# Перебираем все параметры
for i, (n_estimator, max_sample, max_feature) in enumerate(zip(n_estimators, max_samples, max_features)):
    bagging = BaggingRegressor(n_estimators=n_estimator, max_samples=max_sample, max_features=max_feature)
    bagging.fit(X_train, y_train)
    
    # Оценка на тестовой выборке
    test_score = np.sqrt(mean_squared_error(y_test, bagging.predict(X_test)))
    test_scores[i] = test_score
    
    # Оценка на контрольной выборке
    train_score = np.sqrt(mean_squared_error(y_train, bagging.predict(X_train)))
    train_scores[i] = train_score
    
    # Оценка на перекрестной проверке
    cv_score = np.abs(cross_val_score(bagging, X_train, y_train, cv=k_fold, scoring='neg_mean_squared_error').mean())
    cv_scores[i] = cv_score
plt.figure(figsize=(12, 6))


plt.plot(n_estimators, test_scores, marker='o', label='Test Set')
plt.plot(n_estimators, train_scores, marker='o', label='Train Set')
plt.plot(n_estimators, cv_scores, marker='o', label='Cross Validation')

plt.xlabel('N_Estimators')
plt.ylabel('Accuracy')
plt.title('Dependency of Accuracy on N_Estimators')

plt.legend()
plt.grid(True)
plt.show()
```

```python
best_idx = np.argmax(cv_scores)
optimal_n_estimator = n_estimators[best_idx]
optimal_max_sample = max_samples[best_idx]
optimal_max_feature = max_features[best_idx]
print(f"Оптимальные параметры:\n{optimal_n_estimator=}\n{optimal_max_sample=}\n{optimal_max_feature=}")
```

```python
bagging = BaggingRegressor(n_estimators=optimal_n_estimator, max_samples=optimal_max_sample, max_features=optimal_max_feature)
bagging.fit(X_train, y_train)
y_pred = bagging.predict(X_test)
bagging_score = r2_score(y_test, y_pred)
print(f"r2_score на тестовом наборе данных: {bagging_score}")
```
